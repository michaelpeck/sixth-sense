```python
deltas = [
    {'name':'dy_5m','periods': 5},
    {'name':'dy_10m','periods': 10},
    {'name':'dy_30m','periods': 30},
    {'name':'dy_1h','periods': 60},
    {'name':'dy_2h','periods': 120},
    {'name':'dy_3h','periods': 180},
    {'name':'dy_5h','periods': 300},
    {'name':'dy_8h','periods': 480},
    {'name':'dy_10h','periods': 600},
    {'name':'dy_1d','periods': 1440},
    {'name':'dy_2d','periods': 2880},
    {'name':'dy_3d','periods': 4320},
    {'name':'dy_5d','periods': 7200},
    {'name':'dy_10d','periods': 14400},
    {'name':'dy_15d','periods': 21600},
    {'name':'dy_20d','periods': 28800},
    {'name':'dy_30d','periods': 43200},
    {'name':'dy_45d','periods': 64800},
    {'name':'dy_60d','periods': 86400},
    {'name':'dy_90d','periods': 129600},
    {'name':'dy_180d','periods': 259200},
    {'name':'dy_1y','periods': 525600},
]

def populate_deltas(df):
    for delta in deltas:
        df[delta['name']] = df.groupby('asset').open.pct_change(periods=delta['periods'])
```

```python
vols = [
    {'name':'mv_5m','periods': 5},
    {'name':'mv_10m','periods': 10},
    {'name':'mv_30m','periods': 30},
    {'name':'mv_1h','periods': 60},
    {'name':'mv_2h','periods': 120},
    {'name':'mv_3h','periods': 180},
    {'name':'mv_8h','periods': 300},
    {'name':'mv_10h','periods': 480},
    {'name':'mv_10h','periods': 600},
    {'name':'mv_1d','periods': 1440},
    {'name':'mv_2d','periods': 2880},
    {'name':'mv_3d','periods': 4320},
    {'name':'mv_5d','periods': 7200},
    {'name':'mv_10d','periods': 14400},
    {'name':'mv_15d','periods': 21600},
    {'name':'mv_20d','periods': 28800},
    {'name':'mv_30d','periods': 43200},
    {'name':'mv_45d','periods': 64800},
    {'name':'mv_60d','periods': 86400},
    {'name':'mv_90d','periods': 129600},
    {'name':'mv_180d','periods': 259200},
    {'name':'mv_1y','periods': 525600},
]

def populate_avg_volume(df):
    for delta in vols:
        df[delta['name']] = df.groupby('asset').volume.apply(lambda x : x.shift().rolling(delta['periods'],min_periods=delta['periods']).mean().fillna(0))
```

```python
idx = 1
for item in asset_list:
    base_df = item['df']
    # Reset index to start from 0
    df = base_df.reset_index(drop=True)
    # Populate deltas over various time periods
    populate_deltas(df)
    # Populate average volume
    populate_avg_volume(df)
    # df.index.name = 'id'
    item['formatted_df'] = df


```
